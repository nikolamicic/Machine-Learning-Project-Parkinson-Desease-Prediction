```python
# import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load parkinsons dataset into pandas DataFrame

data = pd.read_csv("parkinsons.data")
```

```python
# Correlation heatmap
# New figure with a specific size. The figsize parameter defines the width and height of the figure in inches.
plt.figure(figsize=(12, 8))

# FutureWarning: The default value of numeric_only in DataFrame.corr is deprecated. In a future version, it will default to False. Select only valid columns or specify the value of numeric_only to silence this warning.
# sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
# USE data.corr(numeric_only = True) insted of data.corr()

sns.heatmap(data.corr(numeric_only = True), annot=True, cmap="coolwarm")
# This line creates a heatmap using Seaborn's heatmap function. The function takes several parameters:
# data.corr(): This computes the correlation matrix of the numeric columns in the DataFrame data.
# annot=True: This parameter adds the correlation values to the cells of the heatmap.
# cmap="coolwarm": This parameter specifies the color map to use for the heatmap. "coolwarm" is a diverging color map that ranges from cool (blue) to warm (red), with white in the middle.

# The resulting heatmap provides a visual representation of the correlations between pairs of numerical features 
# in your dataset. The color of each cell in the heatmap indicates the strength and direction of the correlation 
# between the corresponding features. Positive correlations are represented by warmer colors (reds), 
# negative correlations by cooler colors (blues), and no correlation by white. 
# The annot=True parameter adds the actual correlation values to each cell for more detailed information.

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.title('Distribution of Target Variable')
plt.xticks([0,1])
plt.xlabel('0 - healthy, 1 - unhealthy')
plt.hist([data['status']], color=['cadetblue'])
plt.show()
# The result provides a visual representation of how many individuals are classified as 
# healthy (status=0) and how many are classified as unhealthy (status=1) in your dataset. 
# The x-axis shows the different status categories, and the height of each bar represents the count of 
# occurrences for each category. This type of plot is particularly useful for understanding the class 
# distribution in classification tasks.
```

```python
# Histogram of all features
data.hist(figsize = [15,15])
```

```python
# Histograms of selected features
selected_features = ['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)']
data[selected_features].hist(bins=20, figsize=(12, 8))
plt.suptitle("Feature Distribution")
plt.show()

# The resulting output will show a grid of histograms, each corresponding to one of the selected features. 
# The x-axis of each histogram represents the values of the respective feature, and the y-axis represents 
# the frequency of occurrences of those values. The bins parameter controls the number of bars or bins in each 
# histogram. The figsize parameter controls the size of the overall figure that contains the histograms.
```

```python
# Histograms of selected features
selected_features = ['status']
data[selected_features].hist(bins=20, figsize=(12, 8))
plt.suptitle("Feature Distribution")
plt.show()
```

```python
# Box plots for selected features
selected_features = ['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)']
plt.figure(figsize=(12, 8))
sns.boxplot(data=data[selected_features])
plt.title("Box Plots of Selected Features")
plt.xticks(rotation=45)
plt.show()

# The resulting box plots provide insights into the distribution of the selected features. 
# Each box plot displays the median (middle line of the box), interquartile range (IQR, the box), 
# and any potential outliers (individual points outside the whiskers). 
# This visualization helps you understand the central tendency, spread, and presence of outliers in the 
# data for each selected feature.
```
